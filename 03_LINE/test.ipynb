```python
import pandas as pd 
import numpy as np 
import networkx as nx 
```

```python
nodes = pd.read_csv("dataset/Flickr/filtered/users.csv",names=["node"])
edges = pd.read_csv("dataset/Flickr/filtered/links.csv",names=["node1","node2"])
```

```python
edges.values.tolist()
```

```python
G = nx.Graph()
G.add_nodes_from(nodes["node"].values.tolist())
G.add_edges_from(edges.values.tolist())
```

```python
# 根据节点的度进行采样
nodes2,degrees = zip(*G.degree())
nodes2
```

```python
sample_prob = np.array(degrees)
sample_prob = sample_prob / sample_prob.sum()
sample_prob
```

```python
# 将边的权重设置成两个节点度的均值
for u,v in G.edges():
    degree_u = G.degree[u]
    degree_v = G.degree[v]
    G[u][v]['weight'] = (degree_u + degree_v) / 2 
    
```

```python
edges_and_weight = [ (u,v, data["weight"]) for u,v,data in G.edges(data=True)]
edges_and_weight
```

```python
edge_weights = [s[2] for s in edges_and_weight]
edge_weights = np.array(edge_weights)
edge_weights = edge_weights/edge_weights.sum()
edge_weights
```

```python
# 获取采样表
table_size = 1e7
edge_sampling_table = []
for index, weight in enumerate(edge_weights):
    weight
    edge_sampling_table.extend([index]*int(table_size*weight))
```

```python
# 获取正样本
pos_index = np.random.choice(np.arange(0,len(edge_weights)),p=edge_weights)
pos_index
```

```python
sorece_node = edges_and_weight[pos_index][0]
traget_node = edges_and_weight[pos_index][1]
sorece_node, traget_node
```
