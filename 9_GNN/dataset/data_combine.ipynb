```python
import pandas as pd
import torch
import numpy as np
import json
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader
from torch.optim import Adam
import torch.nn as nn
from torch.nn import functional as F
```

```python
atom_188 = pd.read_csv(
    "Mutagenesis-188/atoms.csv",
    delimiter=";",
)
bond_188 = pd.read_csv("Mutagenesis-188/bonds.csv", delimiter=";")
molecule_188 = pd.read_csv("Mutagenesis-188/drugs.csv", delimiter=";")
```

```python
atom_42 = pd.read_csv(
    "Mutagenesis-42/atoms.csv",
    delimiter=";",
)
bond_42 = pd.read_csv("Mutagenesis-42/bonds.csv", delimiter=";")
molecule_42 = pd.read_csv("Mutagenesis-42/drugs.csv", delimiter=";")
```

```python
atom_id_base = atom_188["id"].max() + 1
drug_id_base = molecule_188["id"].max() + 1
bond_id_base = bond_188["id"].max() + 1
```

```python
atom_42_add = atom_42.copy()
molecule_42_add = molecule_42.copy()
bond_42_add = bond_42.copy()
atom_42_add["id"] += atom_id_base
atom_42_add["drug_id"] += drug_id_base
molecule_42_add["id"] += drug_id_base
bond_42_add["id"] += bond_id_base
bond_42_add["drug_id"] += drug_id_base
bond_42_add["atom1_id"] += atom_id_base
bond_42_add["atom2_id"] += atom_id_base
```

```python
atom_all = pd.concat([atom_188,atom_42_add],axis=0)
molecule_all = pd.concat([molecule_188,molecule_42_add],axis=0)
bond_all = pd.concat([bond_188,bond_42_add],axis=0)
```

```python
atom_all.to_csv("Mutagenesis-230/atoms.csv",index=False,sep=";")
molecule_all.to_csv("Mutagenesis-230/drugs.csv",index=False,sep=";")
bond_all.to_csv("Mutagenesis-230/bonds.csv",index=False,sep=";")
```
